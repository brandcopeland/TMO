```python
from sklearn.datasets import load_wine
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загружаем данные
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
# Проверим наличие пропусков
print(df.isnull().sum())
```

```python
# 1. Гистограммы признаков
# Они помогают понять распределение значений, наличие выбросов и асимметрию.
df.hist(bins=15, figsize=(15, 10))
plt.suptitle('Распределение признаков')
plt.show()
```

```python
# 2. Boxplot
# Boxplot показывает медиану, квартили и выбросы. Хорошо для сравнения масштабов и выявления выбросов.
plt.figure(figsize=(15, 8))
sns.boxplot(data=df.drop('target', axis=1), orient='h')
plt.title('Boxplot по всем признакам')
plt.show()
```

```python
# 3. Корреляционная матрица
# Позволяет выявить признаки, которые сильно коррелируют друг с другом.
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Корреляционная матрица признаков')
plt.show()
```

```python
# Гистограмма произвольной колонки
plt.figure(figsize=(8, 5))
sns.histplot(df["alcohol"], bins=20, kde=True)
plt.title('Гистограмма alcohol')
plt.xlabel('Alcohol')
plt.ylabel('Частота')
plt.show()
```

## Вывод

1. **Отсутствие пропусков**  
   В датасете нет пропущенных значений

2. **Гистограммы признаков**  
   - Признаки имеют разные распределения: нормальные (`alcohol`) и с выбросами (`proline`).

3. **Boxplot**  
   - Обнаружены выбросы в признаках: `proline`, `od280/od315_of_diluted_wines`, `color_intensity`.

4. **Корреляционная матрица**  
   - Сильная положительная корреляция между:
     - `flavanoids` и `od280/od315_of_diluted_wines`;
     - `alcohol` и `proline`.

5. **Диаграммы рассеяния (pairplot)**  
   - Классы хорошо разделяются по признакам: `alcohol`, `color_intensity`, `flavanoids`.

6. **Гистограмма `alcohol`**  
   - Распределение близко к нормальному, максимум — около 13.
